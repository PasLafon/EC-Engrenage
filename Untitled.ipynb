```python
# impot de la bibliotheque de calcul
import numpy as np
# impot de la bibliotheque de tracé
import matplotlib.pyplot as plt
```

```python
# Calcul du module minimal pour passer le couple nécessaire avec les contrainte géométriques nécessaire dans le plus petit espace
# définition du nombre minimum de dents
i=1/4
# définition du nombre minimum de dents
Z_min=18
# définition du couple d'entrée
Ce=500
# définition de la contrainte admissible
sigma_adm=70
# définition de k (b=k.m)
k=10
# calcul du module minimum pour passer le couple Ce avec Z_min dents et un ratio de forme k
m_min=pow(11*Ce/(Z_min*sigma_adm*k),1/3)
print('Le module minimum vaut %2.3f mm'%(m_min))
```

```python
# vérification que la contrainte admissible est supportable
m=m_min
Ze=Z_min
Zs=Ze/i
a=m*(Ze+Zs)/2
h=m*(Ze+Zs)
b=k*m
d=m*Z
T=2*Ce/d
sigma=5.5*T/(k*pow(m,2))
print('La contrainte au pied de la dent vaut %3.3f MPa'%(sigma))
print('L encombrment de l engrenage vaut %3.3f mm de hauteur et %3.3f mm de largeur'%(h,b))
```

```python
modules = np.array([.25,.3,.4,.5,.6,.75,.8,1,1.25,1.5,1.75,2])
largeurs = np.array([2.5,2.5,3,5,6,6,8,8,8,12,18,20])
Z_mini=np.ceil(11*Ce/(largeurs*modules*sigma_adm))
hauteurs=modules*Z_mini*(1+1/i)
plt.plot(modules,Z_mini)
plt.plot(modules,hauteurs)
print('Les nombres de dents',Z_mini)
print('Les hauteurs ',hauteurs)
```

```python
Z_mini[Z_mini<Z_min]=Z_min
hauteurs=modules*Z_mini*(1+1/i)
plt.plot(modules,Z_mini)
plt.plot(modules,hauteurs)
print('Les modules',modules)
print('Les nombres de dents',Z_mini)
print('Les hauteurs ',hauteurs)
print('La hauteur la plus faible est de %3.3f mm pour un module de %2.2f mm et un pignon d entrée de %3.0f dents.'%(np.min(hauteurs),modules[np.argmin(hauteurs)],Z_mini[np.argmin(hauteurs)]))
```
